```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import json
import numpy as np
```

```python
df = pd.read_csv('analystt.csv')
```

```python
df.drop(['web-scraper-start-url'],axis=1,inplace=True)
df.rename(columns={'ASIN':'Product Name'},inplace=True)
```

```python
df.drop(['Product Details-ASIN'],axis=1,inplace=True)
```

```python
df['ASIN']=np.nan
df['Manufacturer']=np.nan
```

```python
for idx,row in df.iterrows():
    print(idx)
    data = df.iloc[idx]['Product Details'].replace('\n', '')
    try:
        data_dict = json.loads(data)[0]
    except:
        continue
    product_details=data_dict.get('Product Details')
    cleaned_product_details = product_details.replace('\n', '').replace(' ', '')
    cleaned_product_details = cleaned_product_details.replace('\u200B', '')
    cleaned_product_details = cleaned_product_details.replace('\u200e', '')
    cleaned_product_details = cleaned_product_details.replace('\u200f', '')
    asin_index = cleaned_product_details.find("ASIN:")
    if asin_index != -1:
        # Extract the substring after "ASIN:"
        asin_substring = cleaned_product_details[asin_index + len("ASIN:"):]
        # Extract the first 10 alphanumeric characters from the ASIN substring
        asin = ''.join(filter(str.isalnum, asin_substring))[:10]
        df.at[idx,'ASIN']=asin
        print(asin)
    manufacturer_index = cleaned_product_details.find("Manufacturer:")
#     manufacturer_index = cleaned_product_details.find("Manufacturer:")
    if manufacturer_index != -1:
        # Extract the substring after "Manufacturer:"
        manufacturer_info = cleaned_product_details[manufacturer_index + len("Manufacturer:"):]

        # Find the index of "ASIN" in the manufacturer info
        asin_index = manufacturer_info.find("ASIN")

        if asin_index != -1:
            # Remove the "ASIN" part if it exists
            manufacturer_info = manufacturer_info[:asin_index]
            
        # Strip any extra characters from the beginning and end
        manufacturer = manufacturer_info.strip(':,. ')
        manufacturer_index = manufacturer.find('Manufacturer')
        if(manufacturer_index==-1):
            print(manufacturer)
            df.at[idx,'Manufacturer']=manufacturer
            continue
        else:
            manufacturer_info = manufacturer[manufacturer_index + len("Manufacturer:"):]

            # Find the index of "ASIN" in the manufacturer info
            asin_index = manufacturer_info.find("ASIN")

            if asin_index != -1:
                # Remove the "ASIN" part if it exists
                manufacturer_info = manufacturer_info[:asin_index]

            # Strip any extra characters from the beginning and end
            manufacturer = manufacturer_info.strip(':,. ')
        print(manufacturer)
        df.at[idx,'Manufacturer']=manufacturer
            
    else:
        print("Manufacturer not found in product details")

    
    
```
